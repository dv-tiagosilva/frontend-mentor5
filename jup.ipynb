````python
import numpy as np
from matplotlib import pyplot as plt
import math
````

````python
x = np.linspace(0, 2*3.14, 200)
y = np.zeros(200)
c = np.zeros(200)

for i in range(200):
    y[i] = 5 * math.sin(10*x[i] + 3.14/2) + 5
    
for i in range(200):
    c[i] = 5* math.cos(x[i]) + 5
````

````python
plt.plot(x, y)
plt.plot(x, c)
plt.show()
````
